```python
from scipy.stats import norm
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
#create a histogram
sns.distplot(index_2017['GDP Growth Rate'].dropna()) 
plt.title('GDP Growth Rate 2017')
plt.show()
```

```python
meanP=index_2017['GDP Growth Rate'].mean()
mdP=index_2017['GDP Growth Rate'].median()

sns.distplot(index_2017['GDP Growth Rate'].dropna())
plt.title('GDP Growth Rate 2017')
plt.axvline(meanP, color='b', linestyle='dashed', linewidth=1,label='mean')
plt.axvline(mdP, color='r', linestyle='dashed', linewidth=1,label='median')
plt.show()
```

```python
Chad=index_2017[index_2017['Country Name']=='Chad']
El_Salvador=index_2017[index_2017["Country Name"]=='El Salvador']
```

```python
plt.hist(Chad['GDP Growth Rate'], color='pink',label='Chad')
plt.hist(El_Salvador['GDP Growth Rate'], color='green',label='El Salvador')
plt.show()
```
